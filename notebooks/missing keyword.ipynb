```python
import spacy
nlp=spacy.load('en_core_web_sm')
```

```python
def find(des:str)->list:
    keywords=[];
    doc=nlp(des)
    for chunk in doc.noun_chunks:
        # print(chunk.pos_)
        # if chunk.pos_ in ['NOUN','PROPN'] and not chunk.is_stop:
             # print(token.text)
        keywords.append(chunk.text.lower())
        # pass
    return keywords
    
```

```python
print(find('Machine learning enginner with python sql and aws'))
```

```python
def missing_skills(jd:str,resume:str)->list:
    jd_skills=find(jd)
    resume_skills=find(resume)

    missing=[]

    for ele in jd_skills:
        if ele not in resume_skills:
            missing.append(ele);

    return f'Missing skills: {missing}'
```

```python
missing_skills('Machine learning enginner with AI,aws,GCP,big data,sql,linux','Machine learning enginner with python sql and aws')
```
