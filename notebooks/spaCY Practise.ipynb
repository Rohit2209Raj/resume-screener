```python
import spacy
```

```python
nlp=spacy.load('en_core_web_sm')
```

```python
text = 'I am Rohit Raj and I am living in Sonpur,Bihar'
```

```python
doc=nlp(text)
```

```python
for token in doc:
    print(token.text)
```

```python
for token in doc:
    print(token.lemma_)
```

```python
for token in doc:
    print(token.text+"   ->   "+token.pos_)
```

```python
for token in doc:
    if(token.is_stop):
        print(token.text)
```

```python
for token in doc:
    if(token.is_punct):
        print(token.text)
```

```python
for token in doc.ents:
    print(token.label_)
```
