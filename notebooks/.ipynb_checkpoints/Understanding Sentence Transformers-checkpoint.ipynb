```python
from sentence_transformers import SentenceTransformer,util
```

```python
model=SentenceTransformer('all-MiniLM-L6-v2')
```

```python
sentences = ['This framework generates embeddings for each input sentence',
    'Sentences are passed as a list of string.']
```

```python
embeddings=model.encode(sentences)
```

```python
for sentence,embedding in zip(sentences,embeddings):
    print("Sentence: ",sentence)
    print("Embedding: ",embedding)
    
```

# Cosine - Similarity

```python
# emb1=model.encode(sentences[0])
# emb2=model.encode(sentences[1])

emb1=model.encode('I love Rohit')
emb2=model.encode('Rohit is my love')
similarity=util.cos_sim(emb1,emb2)
print("Similarty: ",similarity*100)
```

# Compute cosine similarity between all pairs

```python

# Compute cosine similarity between all pairs

sentences = ['A man is eating food.',
          'A man is eating a piece of bread.',
          'The girl is carrying a baby.',
          'A man is riding a horse.',
          'A woman is playing violin.',
          'Two men pushed carts through the woods.',
          'A man is riding a white horse on an enclosed ground.',
          'A monkey is playing drums.',
          'Someone in a gorilla costume is playing a set of drums.'
          ]
embedding=model.encode(sentences)
sim=util.cos_sim(embedding,embedding)
sim
```

```python
check=[]
for i in range (0,len(sentences)):
    for j in range (i+1,len(sentences)):
        check.append(((sim[i][j]),i,j))
check
```

```python
sorted_order=sorted(check,key=lambda x:x[0],reverse=True)
sorted_order
```

```python
for score,i,j in sorted_order:
    print('{} ----> {} ---> {:.4f}'.format(sentences[i],sentences[j],score))
```

# Semantic Search

```python
from sentence_transformers import SentenceTransformer,util
```

```python
model=SentenceTransformer('clips/mfaq')
```

```python
question = "How many models can I host on HuggingFace?"
answer_1 = "HuggingFace allows max 5 model hosting per user."
answer_2 = "AutoNLP is an automatic way to train and deploy state-of-the-art NLP models, seamlessly integrated with the Hugging Face ecosystem."
answer_3 = "Based on how much training data and model variants are created, we send you a compute cost and payment link - as low as $10 per job."
```

```python
query_embedding=model.encode(question)
```

```python
corpus_embedding=model.encode([answer_1,answer_2,answer_3])
```

```python
print(util.semantic_search(query_embedding,corpus_embedding))
```
